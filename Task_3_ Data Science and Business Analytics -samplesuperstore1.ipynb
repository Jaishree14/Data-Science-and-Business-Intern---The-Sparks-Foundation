```python
# dropping the duplicates
dataset = dataset.drop_duplicates()
```

```python
#removing the unnecessary columns such as postal code
dataset = dataset.drop(['Postal Code'],axis=1)
```

****Exploring the dataset****

```python
# visualizing using the pair plot
import seaborn as sns 
sns.pairplot(dataset)
```

```python
# correlations between the columns and analyzing using heatmaps
dataset.corr()
plt.figure(figsize=(10,5))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

```python
#categories
plt.figure(figsize = (7,7))
textprops = {"fontsize":18}
plt.title('Category')
plt.pie(dataset['Category'].value_counts(), labels=dataset['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

```python
# computing top categories in terms of sales from first 500 observations
top_category_s = dataset.groupby("Category").Sales.sum().nlargest(n=500)
# computing top categories in terms of profit from first 500 observations
top_category_p = dataset.groupby("Category").Profit.sum().nlargest(n=500)

# plotting to see it visually
plt.style.use('seaborn')
top_category_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='red')
top_category_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='yellow')
plt.xlabel('Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Category Sales vs Profit",fontsize = 15)
plt.show()
```

```python
# computing top sub-categories in terms of sales from first 500 observations
top_subcategory_s = dataset.groupby("Sub-Category").Sales.sum().nlargest(n = 500)
# computing top sub-categories in terms of profit from first 500 observations
top_subcategory_p = dataset.groupby("Sub-Category").Profit.sum().nlargest(n = 500)

# plotting to see it visually
plt.style.use('seaborn')
top_subcategory_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
top_subcategory_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'yellow')
plt.xlabel('Sub-Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Sub-Category Sales vs Profit",fontsize = 15)
plt.show()
```

```python
#visualizing sales in different regions
plt.figure(figsize = (7,7))
plt.title('Regions')
plt.pie(dataset['Region'].value_counts(), labels=dataset['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

Most sales are from west region.

```python
# computing top states in terms of sales from first 15 observations
top_states_s = dataset.groupby("State").Sales.sum().nlargest(n=15)

# computing top states in terms of profit from first 15 observations
top_states_p = dataset.groupby("State").Profit.sum().nlargest(n = 15)

plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'yellow')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 15 states Sales vs Profit",fontsize = 15)
plt.show()
```
